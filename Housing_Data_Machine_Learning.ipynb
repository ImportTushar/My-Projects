```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

count = 0
num = 2
prime_numbers = []

while count < 10:
    if is_prime(num):
        prime_numbers.append(num)
        count += 1
    num += 1

mean = sum(prime_numbers) / len(prime_numbers)

print("Prime numbers:", prime_numbers)
print("Mean of prime numbers:", mean)
```

```python
Housing['Luxury home'] = 0
for i in range(len(Housing)):
    count = 0
    if Housing['Waterfront View'][i] == 'yes':
        count = count +1
    
    if Housing['Condition of the House'][i] =={'Good','excellent'}:
        count =  count+1
        
    if Housing['Overall Grade'][i] >= 8:
        count = count+1
        
    if count >= 2:
        Housing['Luxury home'][i] = 'yes'
        
    else:
        Housing['Luxury home'][i] = 'No'
        

Housing['Luxury home'].unique()
        
    
    
```

```python
def Luxury_home(row):
    count = 0
    if row[0] == 'yes':
        count = count +1
    
    if row[1] in ['Good', 'Excellent']:
        count = count+1
        
    if row[2] >= 8:
        count = count +1
        
Housing['Luxury_home'] = Housing[['Waterfront View' , 'Condition of the House' , 'Overall Grade']].apply(Luxury_home , axis = 1)
Housing['Luxury_home'].unique()
```
